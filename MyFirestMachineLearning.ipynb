```python
import pandas as pd
url1 = "https://drive.google.com/file/d/1ClVidXIRtsoQTfdGL5e9v5UeHjH0vo5-/view?usp=sharing"
url2 = "https://raw.githubusercontent.com/timothypesi/Data-Sets-For-Machine-Learning-/main/california_housing_train.csv"
df = pd.read_csv(url2)
df
```


#**Data Preparation**

```python
x = df["median_house_value"] # selet the colomn of  median_house_value
x
```

```python
y = df.drop("median_house_value",axis =1) # selet all except the colomn of  median_house_value
y
```

###*split the data as training and testing sets*

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.2, random_state= 100)
```

#train the model**

## *Linear Regression*

```python
from sklearn.linear_model import LinearRegression
x_train = x_train.values.reshape(-1, 1)
x_test = x_test.values.reshape(-1, 1)
lr = LinearRegression()
lr.fit(x_train, y_train)
```

## apply the model for pridiction

```python
y_lr_train_prid =lr.predict(x_train)
y_lr_test_prid =lr.predict(x_test)
```

#**model performance**
compare the y raining data with its pridicted values to check the model performance

```python
from sklearn.metrics import mean_squared_error,r2_score
lr_train_mse = mean_squared_error(y_train, y_lr_train_prid)
lr_train_r2 = r2_score(y_train, y_lr_train_prid)
lr_test_mse = mean_squared_error(y_test, y_lr_test_prid)
lr_test_r2 = r2_score(y_test, y_lr_test_prid)
```

# ***Random forest model ***

## train the model

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(max_depth=2,random_state= 100)
rf.fit(x_train, y_train)
rf
```

#Apply the model to make pridiction

```python
y_rf_train_prid =rf.predict(x_train)
y_rf_test_prid =rf.predict(x_test)

y_rf_train_prid
y_rf_test_prid
```

#model performance

```python
from sklearn.metrics import mean_squared_error,r2_score
rf_train_mse = mean_squared_error(y_train, y_rf_train_prid)
rf_train_r2 = r2_score(y_train, y_rf_train_prid)
rf_test_mse = mean_squared_error(y_test, y_rf_test_prid)
rf_test_r2 = r2_score(y_test, y_rf_test_prid)
```

```python
rf_result = pd.DataFrame(["Random fForest",rf_train_mse,rf_train_r2,rf_test_mse,rf_test_r2]).transpose()
rf_result.columns = ["method","training MSE", "training R2","Test MSE","Test R2"]
rf_result
```
